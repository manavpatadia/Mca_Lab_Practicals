```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#create an empty graph

G = nx.Graph()
G.add_node("spam")
G.add_edge(6,2)  #2 nodes will be created and an edge between them is created
G.add_edge("spam",2)
print(G.nodes())
print(G.edges())
```

```python

nx.draw(G,with_labels=True,font_weight='bold')
plt.show()
```

```python
#various ways of adding nodes to a graph

G = nx.Graph()
G.add_node("V1")
G.add_nodes_from(["V2","V3","V4"])
G.add_nodes_from("VXU")
G.add_nodes_from('ABC')
G.add_node("V8")
H=nx.path_graph(3)
G.add_nodes_from(H)
print(G.nodes())
nx.draw(G,with_labels=True, font_weight='bold')
plt.show()

```

```python
print(len(G))  #print the number of nodes in a graph
```

```python
#various ways of adding edges to a graph
K = nx.Graph([('a','c'),('b','d'),('a','d')])
K.add_edge('b','c')
K.add_edge('a','a') #self loop
K.add_edges_from([('a','b'),('b','e')])
print(K.nodes())
print(K.edges())
```

```python
import matplotlib.pyplot as plt
nx.draw(K,with_labels=True,font_weight='bold')
plt.show()
```

```python
print(list(nx.selfloop_edges(K,data=True)))   #show the self loop edges
```

```python
print(K["a"]) #get the edges incident on node 'a'
```

```python
nx.draw(K,with_labels=True,font_weight='bold')
plt.show()
```

```python
#print adjacency list of the graph
print(K.adj)
```

```python
#adding attributes to node

A = nx.Graph()
A.add_node("V1",name="Sandhya",age=20)
A.add_nodes_from([("V2",{"name":"Adarsh","age":17}),("V3",{"name":"Shreyas","age":12})])
print(A.nodes())
A.nodes["V3"]  #to get all the attributes of V3 node
A.nodes["V3"]["age"] #to get specific attributes of V3 node
# nx.draw(A,with_labels=True,font_weight='bold')
# plt.show()
```

```python
A.add_edges_from([('V1','V2'),('b','e')])
A.nodes["V3"]["age"] #to get specific attributes of V3 node
nx.draw(A,with_labels=True,font_weight='bold')
plt.show()
```

```python
#Adding weights to edges

A.add_edge('e','V1',w=4)
nbr_lst = A.neighbors('e')
print(nbr_lst)
nx.draw(A,with_labels=True,font_weight='bold')
plt.show()
```

```python
A.add_edges_from([('b','V1'),('V2','V3')], w=3)
print(A.edges())
```

```python
#various properties

print(len(A)) #no. of nodes in A
print(A.order()) #no. of nodes in A
print(A.number_of_nodes())  #no. of nodes in A
print(A.number_of_edges()) #get number of edges
print(A.has_node('V1'))  #check if node V1 exists
print(A.has_edge('b','c'))  #check if edge b,c exists
```
