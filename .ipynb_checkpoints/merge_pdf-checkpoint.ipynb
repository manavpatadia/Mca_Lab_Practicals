```python
import fitz
import pandas as pd
import os
```

```python
input_book_file = "Discrete_Mathematics.pdf"
input_odd_book = "Odd_Solutions.pdf"
input_even_book = "Even_Solutions.pdf"

output_book_file = "Rearranged_book.pdf"
output_odd_file = "Rearranged_odd.pdf"
output_even_file = "Rearranged_even.pdf"

book_index = "Book_index.csv"
```

```python
df = pd.read_csv(book_index)
df
```

```python
final_book = fitz.open("final_book.pdf")
for row in df.index:
    file_handle_book = fitz.open(input_book_file)
    file_handle_odd = fitz.open(input_odd_book)
    file_handle_even = fitz.open(input_even_book)
    list_book = list(range(df['B_Page_start'][row]-1, df['B_Page_end'][row]))
    list_odd = list(range(df['O_Page_start'][row]-1, df['O_Page_end'][row]))
    list_even = list(range(df['E_Page_start'][row]-1, df['E_Page_end'][row]))
    print(row, list_book, "\t", list_odd, "\t", list_even)
    file_handle_book.select(list_book)
    file_handle_odd.select(list_odd)
    file_handle_even.select(list_even)
    final_book.insert_pdf(file_handle_book)
    final_book.insert_pdf(file_handle_odd)
    final_book.insert_pdf(file_handle_even)
    file_handle_book.close()
    file_handle_odd.close()
    file_handle_even.close()

```

```python
final_book.save("final_book_new.pdf")
final_book.close()
```
