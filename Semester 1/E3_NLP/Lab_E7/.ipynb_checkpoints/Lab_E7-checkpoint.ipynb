```python
from bs4 import BeautifulSoup
import re
import pandas as pd
import requests
```

```python
url = "https://www.brainyquote.com"
a_c = requests.get(url)
```

```python
#print(a_c.content)
bs = BeautifulSoup(a_c.content)
#print(bs.prettify())
```

```python
quote_cls = bs.find_all('div', attrs={"class":"bqLn"})
```

```python
quote_cls = [q.find_all('a', href=True)[0]['href'] for q in quote_cls] #finding all possible topic links to browse
quote_cls = [url+q for q in quote_cls]
quote_cls[:10]
```

#### <font color='orange'> Extracting Quote, Author and Topic from all Type of Quotes in website

```python
def extract_quote_author(topic_url, topic):
    topic_quote_bs = BeautifulSoup(requests.get(topic_url).content)
    topic_quote_cls_list = topic_quote_bs.find_all('div', attrs={"class":"grid-item qb clearfix bqQt"}) #extracting quote and author text
    topic_quote_author_list = [q.text.strip().split("\n\n\n\n") for q in topic_quote_cls_list] #creating list of quote and author 
    topic_quote_author_list = [q + [topic] for q in topic_quote_author_list] #adding topic to each quote and author
    return topic_quote_author_list
```

```python
all_type_quote_list = []
for topic_url in quote_cls:
    topic = topic_url.split("/")[-1].replace("-quotes","")
    temp_list = extract_quote_author(topic_url, topic)
    all_type_quote_list.extend(temp_list)

all_type_quote_list[:5]
```

```python
df_all_quotes = pd.DataFrame(all_type_quote_list, columns = ['quote', 'author','type'])
```

```python
df_all_quotes.head()
```

```python
df_all_quotes.sample(20)
```

```python
df_all_quotes.to_csv("All_Quotes_Library1.csv", index = False)
```

## <font color='orange'> 2. Use scikit-learn and createfollowing text represenationson the dataset:<br> <dd> a. BoWs <br> b. ngram vectorization <br> c. tf-idf vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
BoW_vectorizer = CountVectorizer(stop_words= 'english')
```

```python
list_all_quotes = list(df_all_quotes['quote'])
list_all_quotes[:2]
```

```python
# fit() function will learn a vocabulary from one or more documents.
BoW_vectorizer.fit(list_all_quotes)
```

```python
#BoW_vectorizer.vocabulary_
```

```python
voc_all_quotes = BoW_vectorizer.get_feature_names()
voc_all_quotes[:10]
```

```python
# transform() function will transform one or more documents to vectors
print(list_all_quotes[:2])
vector = BoW_vectorizer.transform(list_all_quotes)
#print(vector)
print(type(vector))
print(vector.toarray())
```

```python
df_BoW_vectors = pd.DataFrame(vector.toarray(), columns = voc_all_quotes)
```

### <font color='orange'> Ngrams Vectorized

```python
NgramVectorizer = CountVectorizer(stop_words = 'english', ngram_range = (1,3), min_df = 2)
```

```python
NgramVectorizer.fit(list_all_quotes)
```

```python
Ngram_vectors = NgramVectorizer.get_feature_names()
print(Ngram_vectors[:20])
```

```python
vector = NgramVectorizer.transform(list_all_quotes)
print(vector.toarray())
```

```python
df_ngram = pd.DataFrame(data = vector.toarray(), columns = Ngram_vectors)
```

```python
df_ngram
```

### <font color='orange'> TfidfVectorizer
