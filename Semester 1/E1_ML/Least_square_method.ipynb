```python
import numpy as np
from matplotlib import pyplot
x = np.array([0.2343,0.259426,0.258054,0.284599,0.328975,0.346999,0.365385,0.363112,0.3975,0.41918,0.442769,0.444546,0.482704,0.502601,0.518173,0.554894])
y = np.array([60.323,61.122,60.171,61.187,63.221,63.639,64.989,63.761,66.019,67.857,68.169,66.513,68.655,69.564,69.331,70.551])
X = np.c_[np.ones((x.shape[0], 1)), x]
XT = X.T
XT_X = XT.dot(X)
XT_Y = XT.dot(y)
coef = np.linalg.inv(XT_X).dot(XT_Y)
print("coefficient = ", coef)
ypred = X.dot(coef)
print("Y Pred = ", list(ypred))
pyplot.scatter(x, y, color='red')
pyplot.plot(x, ypred, color='blue')
pyplot.show()
```
