**Qn 1**

Download the mammography dataset given. Implement PCA. Find the first two principal components of the dataset and plot it using scatter plot with different colors for each target.  Dataset is obtained from https://www.openml.org/d/310
What are your interpretations?


```python
from sklearn.decomposition import PCA
from numpy import size,mean
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_csv("C:/Users/HP/Downloads/mammography.csv")
print(data)
data = data.values
```

```python
X = data[:,0:6]
Y = data[:,6]
print("X\n",X)
print("\nY\n",Y)

Y_= []
for i in range(len(Y)):
    if Y[i]=="'-1'":
        Y_.append(-1)
    else:
        Y_.append(1)       
plt.scatter(X[:,0],X[:,1],c=Y_)
plt.show()
```

```python
print("Shape of original data is ",X.shape)

pca_data = PCA(2) #reduced to 2 dimensions
data_projection = pca_data.fit_transform(X)
print("\nProjected data: \n",data_projection)

print("\nShape of projected data is ",data_projection.shape)
```

```python
plt.scatter(data_projection[:,0],data_projection[:,1],c=Y_)
plt.show()
```

```python
print("\nCumulative explained variance ratio")
print(np.sum(pca_data.explained_variance_ratio_)) 
```

```python
inverse_of_projected_data = pca_data.inverse_transform(data_projection)
print(inverse_of_projected_data)
```

```python

print(np.linalg.norm(X))

print(np.linalg.norm(inverse_of_projected_data))
```
