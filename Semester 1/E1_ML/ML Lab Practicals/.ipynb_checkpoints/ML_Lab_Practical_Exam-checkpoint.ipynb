####  <font color='orange'> I have tried two ways to do PCA on given Image <br><br> - Split Image to 3 different sets of Images: Red, Blue, Green and then compress these 3 images using PCA(5) and then merge all 3 compressed Image and reconstruct final Image <br><br> - Apply PCA(5) on given image and recounstruct compressed image

### <font color='orange'>Function for calculating PCA for Red, Blue and Green channel separately

```python
def pca_blue_green_red(df_blue, df_green, df_red, components):
    
    pca_b = PCA(n_components = components)
    pca_b.fit(df_blue)
    trans_pca_b = pca_b.transform(df_blue)

    pca_g = PCA(n_components = components)
    pca_g.fit(df_green)
    trans_pca_g = pca_g.transform(df_green)

    pca_r = PCA(n_components = components)
    pca_r.fit(df_red)
    trans_pca_r = pca_r.transform(df_red)
    
    print("Shape of image vector after reducing it to 5 dimentions", trans_pca_b.shape, trans_pca_g.shape, trans_pca_r.shape)
    print(f"Sum of Explained Variance Ratios for Blue Channel : {sum(pca_b.explained_variance_ratio_)}")
    print(f"Sum of Explained Variance Ratios for Green Channel: {sum(pca_g.explained_variance_ratio_)}")
    print(f"Sum of Explained Variance Ratios for Red Channel  : {sum(pca_r.explained_variance_ratio_)}")
    divider = "\n" + "-"*120 + "\n"
    print(divider)
    fig = plt.figure(figsize = (25, 8)) 

    fig.add_subplot(141)
    plt.title("Blue Channel")
    plt.ylabel('Variation explained')
    plt.xlabel('Eigen Value')
    plt.bar(list(range(1,6)),pca_b.explained_variance_ratio_)

    fig.add_subplot(142)
    plt.title("Green Channel")
    plt.ylabel('Variation explained')
    plt.xlabel('Eigen Value')
    plt.bar(list(range(1,6)),pca_g.explained_variance_ratio_)

    fig.add_subplot(143)
    plt.title("Red Channel")
    plt.ylabel('Variation explained')
    plt.xlabel('Eigen Value')
    plt.bar(list(range(1,6)),pca_r.explained_variance_ratio_)
    plt.show()
    print(divider)
    print("Eigen Values for blue channel\n", pca_b.components_)
    print("Eigen Vectors for blue channel\n", pca_b.explained_variance_)
    print(divider)
    print("Eigen Values for green channel\n", pca_g.components_)
    print("Eigen Vectors for green channel\n", pca_g.explained_variance_)
    print(divider)
    print("Eigen Values for red channel\n", pca_r.components_)
    print("Eigen Vectors for red channel\n", pca_r.explained_variance_)
    print(divider)
    b_arr = pca_b.inverse_transform(trans_pca_b)
    g_arr = pca_g.inverse_transform(trans_pca_g)
    r_arr = pca_r.inverse_transform(trans_pca_r)
    
    print("Shape after reconstruction", b_arr.shape, g_arr.shape, r_arr.shape)
    print(divider)
    img_reduced= (cv2.merge((b_arr, g_arr, r_arr)))
    print(img_reduced.shape)
    
    fig = plt.figure(figsize = (10, 7.2)) 
    fig.add_subplot(121)
    plt.title("Original Image")
    plt.imshow(img)
    
    fig.add_subplot(122)
    plt.title("Reduced Image")
    plt.imshow(img_reduced)
    
    plt.show()
    print(divider)
    total_explained_variance_ratio = (pca_b.explained_variance_ratio_ + pca_g.explained_variance_ratio_ + pca_r.explained_variance_ratio_)/3
    #Reconstruction_Error = np.linalg.norm((img-img_reduced),None)
    Reconstruction_Error = 1-np.sum(total_explained_variance_ratio)
    print("Reconstruction_Error = ", Reconstruction_Error)
```

```python
pca_blue_green_red(df_blue, df_green, df_red, 5)
```

```python
def pca_full_image(df_img, components):
    pca_full = PCA(n_components=5)
    pca_full.fit(df_img)
    trans_pca_full = pca_full.transform(df_img)
    print("Shape of image vector after reducing it to 5 dimentions", trans_pca_full.shape)
    print(f"Sum of Explained Variance Ratios for Blue Channel : {sum(pca_full.explained_variance_ratio_)}")
    divider = "\n" + "-"*120 + "\n"
    print(divider)
    
    fig = plt.figure(figsize = (15, 8)) 
    plt.title("Eigen Value vs Variation Explained")
    plt.ylabel('Variation Explained')
    plt.xlabel('Eigen Value')
    plt.bar(list(range(1,6)), pca_full.explained_variance_ratio_)
    
    #Eigen Values and Eigen Vectors for full image compression
    print("Eigen Values for full channel\n", pca_full.components_)

    print("Eigen Vectors for full channel\n", pca_full.explained_variance_)
    
    full_arr = pca_full.inverse_transform(trans_pca_full)
    reshaped_reduced_array = full_arr.reshape((185, 303, 3))

    fig = plt.figure(figsize = (10, 8)) 
    fig.add_subplot(121)
    plt.title("Original Image")
    
    plt.imshow(img)
    
    fig.add_subplot(122)
    plt.title("Reduced Image")
    
    plt.imshow(reshaped_reduced_array)
    plt.show()
    
    Reconstruction_Error2 = 1-np.sum(pca_full.explained_variance_ratio_)
    print("Reconstruction_Error = ", Reconstruction_Error2)
```

```python
pca_full_image(df_img, 5)
```

### <font color='orange'> <center> d. Plot the graph PC versus reconstruction error for varied number of principal components.

```python
sns.set_style("darkgrid")
plt.figure(facecolor='w')
max_comp=100
start=1
error_record=[]
for i in range(start,max_comp):
    pca = PCA(n_components=i, random_state=42)
    pca2_results = pca.fit_transform(img_r)
    pca2_proj_back=pca.inverse_transform(pca2_results)
    total_loss= 1 - np.sum(pca.explained_variance_ratio_)
    #np.linalg.norm((img_r - pca2_proj_back),None)
    error_record.append(total_loss)
plt.clf()
plt.figure(figsize=(20,10))
plt.title("Reconstruction Error of PCA")
plt.plot(error_record,'r')
plt.ylabel('Reconstruction Error')
plt.xlabel('Number of Components')
plt.xticks(range(len(error_record)), range(start,max_comp), rotation='vertical')
plt.xlim([-1, len(error_record)])
plt.show()
```
